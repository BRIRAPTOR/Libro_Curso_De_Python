## 4.3.- Funciones con Múltiples Parámetros

Ahora armaremos una función usando IF que nos dice cual es la raiz cuadrada más grande de 3 números

```python
import math

def raiz_mayor_de_tres(a, b, c):
    raiz_a = math.sqrt(a)
    raiz_b = math.sqrt(b)
    raiz_c = math.sqrt(c)

    if raiz_a >= raiz_b and raiz_a >= raiz_c:
        return raiz_a
    elif raiz_b >= raiz_a and raiz_b >= raiz_c:
        return raiz_b
    else:
        return raiz_c

print("La raíz cuadrada más grande es:", raiz_mayor_de_tres(9, 16, 25))
```

Por si las dudas, esto se podría haber simplificado usando Max y Min

```python
# Uso de max y min
print("Máximo de la lista:", max([15, 82, 13, 55]))
print("Mínimo de la palabra:", min("amarillo"))
```

Ahora vamos a escribir un código que simule el lanzamiento de una moneda 100,000 veces. No vamos a mostrar 100,000 resultados, solo una pequeña tabla con todas las caras y todas la cruces

```python
import random

# Contadores para caras y cruces
caras = 0
cruces = 0

# Simulación de 100,000 lanzamientos
for _ in range(100000):
    if random.choice([True, False]):
        caras += 1
    else:
        cruces += 1

# Mostrar el resultado
print("Resultados de los 100,000 lanzamientos de moneda:")
print("Caras:", caras)
print("Cruces:", cruces)
```

Si queremos obligar a que random sieeeempre nos de los mismos resultados, podemos usar seed. Arma una semilla de 25 y ejecuta el código 2 veces, checa como se repita (seed necesita repetirse 2 veces tmb)

```python
import random

random.seed(25)

# Contadores para caras y cruces
caras = 0
cruces = 0

# Simulación de 100,000 lanzamientos con semilla
for _ in range(100000):
    if random.choice([True, False]):
        caras += 1
    else:
        cruces += 1

# Mostrar el resultado
print("Resultados de los 100,000 lanzamientos de moneda con semilla:")
print("Caras:", caras)
print("Cruces:", cruces)

# Volver a establecer la semilla para obtener los mismos resultados
random.seed(25)

# Simulación de 100,000 lanzamientos con los mismos resultados
caras2 = 0
cruces2 = 0

for _ in range(100000):
    if random.choice([True, False]):
        caras2 += 1
    else:
        cruces2 += 1

print("\nResultados de los 100,000 lanzamientos de moneda nuevamente con semilla:")
print("Caras:", caras2)
print("Cruces:", cruces2)
```

```python
import random

# Función para lanzar 2 dados
def lanzar_dados():
    dado1 = random.randint(1, 6)
    dado2 = random.randint(1, 6)
    return dado1, dado2

# Función para sumar los resultados de los dados
def suma_dados(dado1, dado2):
    return dado1 + dado2

# Función para jugar "El Gran 8"
def jugar_el_gran_8():
    while True:
        dado1, dado2 = lanzar_dados()  # Lanzamos los dados
        suma = suma_dados(dado1, dado2)  # Sumamos los resultados
        print(f"Lanzamiento de dados: {dado1} y {dado2}, suma: {suma}")

        if suma == 8:
            print("¡Ganaste!")
            break
        elif suma == 7:
            print("Perdiste!")
            break
        else:
            print("Continuamos...")

# Llamamos a la función para comenzar el juego
jugar_el_gran_8()
```

También, hay funciones que pueden aceptar una cantidad indefinida de parametros - arma una función llamada promedio_calif que te diga el promedio de una cantidad arbitraria de calificaciones

```python
def promedio_calif(*calificaciones):
    """Esta función calcula el promedio de las calificaciones"""
    if len(calificaciones) == 0:
        return "No hay calificaciones"
    return sum(calificaciones) / len(calificaciones)

# Ejemplo de uso
print(promedio_calif(80, 90, 95))  # Resultado: 88.33
```

Intenta modificar aprox pi desde una función - verás que python no lo hace, solo modifica una copia

```python
aprox_pi = 3.141  # Variable global

def modificar_pi():
    aprox_pi = 3.14  # Modifica solo la variable local
    print("Valor local de pi:", aprox_pi)

modificar_pi()  # Muestra el valor local
print("Valor global de pi:", aprox_pi)  # Muestra el valor global
```

```python
aprox_pi = 3.141  # Variable global

def modificar_pi():
    global aprox_pi  # Usa la variable global
    aprox_pi = 3.14

modificar_pi()  # Modifica la variable global
print("Valor global de pi:", aprox_pi)  # Muestra el valor global modificado
```

## 4.8.- Medidas de tendencia central y Dispersión

Importa la librería d eestadistica, y crea una lista de calificaciones aleatorias entre 50 y 100

```python
import statistics
import random
```

```python
calificaciones = [random.randint(50, 100) for _ in range(10)]
print("Calificaciones:", calificaciones)
```

```python
media = statistics.mean(calificaciones)
mediana = statistics.median(calificaciones)
moda = statistics.mode(calificaciones)

print("Media:", media)
print("Mediana:", mediana)
print("Moda:", moda)
```

```python
varianza = statistics.variance(calificaciones)
desviacion_estandar = statistics.stdev(calificaciones)

print("Varianza:", varianza)
print("Desviación estándar:", desviacion_estandar)
```
